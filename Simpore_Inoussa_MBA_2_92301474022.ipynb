**Tokenization**

```python

import pdfplumber
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

# Download necessary NLTK data files (only need to run once)
nltk.download('punkt')

# Function to extract text from PDF using pdfplumber
def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"  # Adding newline for better formatting
    return text

# ---  Set the pdf_path to the provided location ---
# To fix the 'FileNotFoundError', please upload the '04MB0440-Design-Thinking.pdf' file
# to your Colab environment, for example, by dragging and dropping it into the files section (folder icon on the left sidebar).
# Ensure the path below matches the location where you uploaded the file.
pdf_path = '/content/04MB0440-Design-Thinking.pdf'


extracted_text = extract_text_from_pdf(pdf_path)

# Tokenization
word_tokens = word_tokenize(extracted_text)
sentence_tokens = sent_tokenize(extracted_text)

# Output the results
print("Word Tokens:")
print(word_tokens[:50])  # Displaying first 50 tokens for brevity
print("\nSentence Tokens:")
print(sentence_tokens[:5])  # Displaying first 5 sentences for brevity# Displaying first 5 sentences for brevityDisplaying first 5 sentences for brevity
```

Stop words **removal**

```python

import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')
def extract_text_from_pdf(pdf_path):
    text = ""

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"
    return text


def remove_stop_words(tokens):
    stop_words = set(stopwords.words('english'))
    return [word for word in tokens if word.lower() not in stop_words]


pdf_path = '/content/04MB0440-Design-Thinking.pdf'
extracted_text = extract_text_from_pdf(pdf_path)
word_tokens = word_tokenize(extracted_text)
filtered_tokens = remove_stop_words(word_tokens)
print("Original Word Tokens:")
print(word_tokens[:50])
print("\nFiltered Word Tokens (Stop Words Removed):")
print(filtered_tokens[:50])
```

3. Stemming of **words**

```python
import pdfplumber
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer


nltk.download('punkt')
nltk.download('stopwords')


def extract_text_from_pdf(pdf_path):
    text = ""

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"
    return text

def remove_stop_words(tokens):
    stop_words = set(stopwords.words('english'))
    return [word for word in tokens if word.lower() not in stop_words]


def stem_words(tokens):
    ps = PorterStemmer()
    return [ps.stem(word) for word in tokens]

pdf_path = '/content/04MB0440-Design-Thinking.pdf'

extracted_text = extract_text_from_pdf(pdf_path)
word_tokens = word_tokenize(extracted_text)
filtered_tokens = remove_stop_words(word_tokens)
stemmed_tokens = stem_words(filtered_tokens)
print("Original Word Tokens:")
print(word_tokens[:50])
print("\nFiltered Word Tokens (Stop Words Removed):")
print(filtered_tokens[:50])
print("\nStemmed Word Tokens:")
print(stemmed_tokens[:50])
```

4. TFID **vectorization**

```python
import pdfplumber
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer

# Download necessary NLTK data files (only need to run once)
nltk.download('punkt')
nltk.download('stopwords')

# Function to extract text from PDF using pdfplumber
def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"  # Adding newline for better formatting
    return text

# Function to remove stop words
def remove_stop_words(tokens):
    stop_words = set(stopwords.words('english'))
    return [word for word in tokens if word.lower() not in stop_words]


def stem_words(tokens):
    ps = PorterStemmer()
    return [ps.stem(word) for word in tokens]

pdf_path = '/content/04MB0440-Design-Thinking.pdf'

extracted_text = extract_text_from_pdf(pdf_path)
word_tokens = word_tokenize(extracted_text)
filtered_tokens = remove_stop_words(word_tokens)

stemmed_tokens = stem_words(filtered_tokens)
preprocessed_text = ' '.join(stemmed_tokens)

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([preprocessed_text])
feature_names = vectorizer.get_feature_names_out()
import pandas as pd
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

print("TF-IDF Matrix:")
print(tfidf_df)
```
